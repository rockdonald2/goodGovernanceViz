```python
resp = requests.get('https://en.wikipedia.org/wiki/Freedom_in_the_World')
soup = BeautifulSoup(resp.content, 'lxml')
```

```python
table = soup.select_one('#mw-content-text > div.mw-parser-output > table:nth-child(18) > tbody')
```

```python
result = []
regexp = re.compile('\*')

# 3 - 198
for n in range(3, 198):
    pr = int(table.select_one('tr:nth-child({}) > td:nth-child(19)'.format(n)).text)
    cl = int(table.select_one('tr:nth-child({}) > td:nth-child(20)'.format(n)).text)
    country = re.sub(regexp, '', table.select_one('tr:nth-child({}) > td:nth-child(1) > a'.format(n)).text)
    result.append({'Country': country, 'Value': (pr + cl) / 2})
```

```python
fr = pd.DataFrame(result)
```

```python
fr = fr.merge(codes, on='Country', how='left')
```

```python
pop = pd.read_csv('./population.csv')
```

```python
pop.rename(columns={'Country Name': 'Country', 'Country Code': 'Code'}, inplace=True)
```

```python
cc = codes.Code.tolist()
```

```python
result = []

for c in cc:
    country = pop[(pop.Code == '{}'.format(c)) & (pop.Year == 2018)].Country
    
    if not country.values.size > 0:
        continue
    
    country = country.values[0]
    code = c
    value = int(pop[(pop.Code == '{}'.format(c)) & (pop.Year == 2018)].Value)
    size = 'big' if value > 35000000 else ('midsize' if 35000000 > value > 5000000 else 'small')
    result.append({'Country': country, 'Code': code, 'Value': value, 'Size': size})
```

```python
with open('../data/pop.json', 'w') as outfile:
    json.dump(result, outfile)
```
